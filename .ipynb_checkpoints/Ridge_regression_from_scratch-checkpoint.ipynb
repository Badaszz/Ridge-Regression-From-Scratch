```python
import numpy as np
import pandas as pd
```

```python
X = np.loadtxt("X_train.csv", delimiter = ",") # features data
y = np.loadtxt("y_train.csv", delimiter = ",") # Target data
```

```python
# Step 1 : Firstly make the features and target data mean centred 
# i.e X_new = X - mean(X_data)  y_new = y - mean(y_data)

# Compute means for centering 
mu_X = X.mean(axis=0)   # vector of length n_features
mu_y = y.mean()         # scalar

# Center data to avoid penalizing the intercept
Xc = X - mu_X           
yc = y - mu_y

n_features = X.shape[1]
```

```python
# Step 2 : find the ridge regression solution for the weights, w
# i.e solve ((X.T * X) + lambda*I)w = X.T * y
# where X is the features, y is the target and lambda is the regularization hyperparameter to penalize large values of w

# let lambda be 0.1
lam = 0.1 # would tune this later

# let A = ((X.T * X) + lambda*I) and b = X.T * y
A = Xc.T @ Xc + lam * np.eye(n_features)  
b = Xc.T @ yc

# find w for Aw = b 
w = np.linalg.solve(A, b)

```

```python
# Step 3 : Reecover the intercept so predictions are in original scale
# intercept = mean(y) - mean(X)*w
# OR use the weights as is then uncentre the prediction y_pred = (X_new - mean_X)*w + mean_y
# i am going with the first option

b = mu_y - mu_X @ w
```

```python
# Step 4 : make predictoin and then evaluate to tune lambda

y_pred = X @ w + b

print(np.mean((y - y_pred) ** 2))
```

```python
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - y.mean()) ** 2)
print( 1.0 - ss_res / ss_tot) if ss_tot > 0 else print(1.0) 
```

```python
# now i would package all of this into functions

## Function to find the weights and bias for the data
def ridge_closed_form(X, y, lam):
    """
    Fit Ridge (L2) regression WITHOUT penalizing the intercept, using the closed form (i.e. the mean centred data for training).

    Parameters
    ----------
    X : array-like, shape (n_samples, n_features) - Feature matrix.
    y : array-like, shape (n_samples,) or (n_samples, 1) - Target vector.
    lam : float - Regularization parameter (lambda >= 0). Larger values => more penalized weights.

    Returns
    -------
    w : ndarray, shape (n_features,) - Fitted weights.
    b : float - Intercept term (not regularized).
    """
    # Ensure NumPy arrays (np.asarray) and numeric dtype
    X = np.asarray(X, dtype=float)          # converts input to NumPy array
    y = np.asarray(y, dtype=float).ravel()  # ravel flattens to 1D if shape was (n,1)

    # Compute means for centering (np.mean with axis=0 means column-wise)
    mu_X = X.mean(axis=0)   # vector of length n_features
    mu_y = y.mean()         # scalar

    # Center data to avoid penalizing the intercept
    Xc = X - mu_X           
    yc = y - mu_y

    n_features = X.shape[1]

    # Build (Xc^T*Xc + lam*I) and RHS Xc^T * yc
    A = Xc.T @ Xc + lam * np.eye(n_features)  # np.eye makes an identity matrix
    b_vec = Xc.T @ yc

    # Solve linear system A w = b 
    w = np.linalg.solve(A, b_vec)

    # Recover intercept so predictions are in original scale
    b = mu_y - mu_X @ w
    return w, b

## Function to make predictions

def predict_ridge(X, w, b):
    """
    Predict using fitted Ridge weights and intercept.

    Parameters
    ----------
    X : array-like, shape (n_samples, n_features)
    w : ndarray, shape (n_features,)
    b : float

    Returns
    -------
    y_hat : ndarray, shape (n_samples,)
    """
    X = np.asarray(X, dtype=float)
    return X @ w + b


## Functions to evaluate the model

def mse(y_true, y_pred):
    """
    Mean Squared Error: average of (y_true - y_pred)^2
    """
    y_true = np.asarray(y_true, dtype=float).ravel()
    y_pred = np.asarray(y_pred, dtype=float).ravel()
    return np.mean((y_true - y_pred) ** 2)


def r2_score(y_true, y_pred):
    """
    R^2 = 1 - SS_res / SS_tot
    """
    y_true = np.asarray(y_true, dtype=float).ravel()
    y_pred = np.asarray(y_pred, dtype=float).ravel()
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - y_true.mean()) ** 2)
    return 1.0 - ss_res / ss_tot if ss_tot > 0 else 1.0

## Functions to select lambda 

def kfold_indices(n_samples, k=5, shuffle=True, random_state=42):
    """
    Generate (train_idx, val_idx) for K-Fold CV using NumPy only.

    - np.arange builds index array [0, 1, 2, ..., n-1]
    - np.random.default_rng creates a random generator for reproducible shuffling
    - rng.shuffle shuffles in-place
    - np.array_split splits indices into k nearly equal folds
    """
    indices = np.arange(n_samples)
    if shuffle:
        rng = np.random.default_rng(random_state)
        rng.shuffle(indices)
    folds = np.array_split(indices, k)
    for i in range(k):
        val_idx = folds[i]
        train_idx = np.concatenate([folds[j] for j in range(k) if j != i])
        yield train_idx, val_idx


def select_lambda_cv(X, y, lambdas, k=5, random_state=42):
    """
    Choose lambda by K-fold CV: lowest mean validation MSE wins.

    Returns
    -------
    best_lam : float
    cv_results : dict mapping lam -> mean_val_mse
    """
    X = np.asarray(X, dtype=float)
    y = np.asarray(y, dtype=float).ravel()
    cv_results = {}
    best_lam, best_score = None, np.inf

    for lam in lambdas:
        fold_mses = []
        for tr_idx, va_idx in kfold_indices(len(X), k=k, shuffle=True, random_state=random_state):
            X_tr, y_tr = X[tr_idx], y[tr_idx]
            X_va, y_va = X[va_idx], y[va_idx]

            w, b = ridge_closed_form(X_tr, y_tr, lam)
            y_hat = predict_ridge(X_va, w, b)
            fold_mses.append(mse(y_va, y_hat))

        mean_mse = float(np.mean(fold_mses))
        cv_results[lam] = mean_mse
        if mean_mse < best_score:
            best_score, best_lam = mean_mse, lam

    return best_lam, cv_results
    
```

```python
## Function to plot the true against the predicted targets

import matplotlib.pyplot as plt
import numpy as np

def plot_true_vs_pred(y_true, y_pred, title="True vs Predicted"):
    """
    Scatter plot of true values vs predicted values with a perfect prediction line.

    Parameters
    ----------
    y_true : array-like
        Ground truth target values.
    y_pred : array-like
        Model predicted values.
    title : str
        Title for the plot.
    """
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    plt.figure(figsize=(6,6))
    plt.scatter(y_true, y_pred, alpha=0.7, label="Predictions")
    plt.xlabel("True y")
    plt.ylabel("Predicted y")
    plt.title(title)

    # Perfect prediction reference line
    plt.plot([y_true.min(), y_true.max()],
             [y_true.min(), y_true.max()],
             'r--', label="Perfect predictions")

    plt.legend()
    plt.show()
```

```python
## lets use these functions

# picking a range of lambdas
lambdas = np.logspace(-2, 2, 10) 

# find th best value for lambda using cross validation k - folds
best_lam, cv_results = select_lambda_cv(X, y, lambdas, k=5, random_state=42)
print("Best lambda from CV:", best_lam)

# use the best lambda for the ridge regression solution
w, b = ridge_closed_form(X, y, best_lam)

# Evaluate on training 
y_hat = predict_ridge(X, w, b)
print("Train MSE:", mse(y, y_hat))
print("Train R^2:", r2_score(y, y_hat))

# make plot of the true and the predicted values
plot_true_vs_pred(y, y_hat, title="True vs Predicted")
```
